```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%matplotlib inline
```

```python
def resident(x, t):
    lamda = (beta1*(x[1]+2*x[2]))
    sdot = rho - mu*x[0] - sigmaS*lamda*x[0]
    i1dot = sigmaS*lamda*x[0] - (mu+alpha1+(sigmaI*lamda))*x[1]
    d11dot = sigmaI*lamda*x[1] - (mu+alpha11)*x[2]
    return sdot,i1dot,d11dot
```

```python
mu = 0.02
rho = 0.02
sigmaS = 1
sigmaI = 5

beta1=0.75
alpha1 = 0.1 - 0.157**2/0.2*(1 - np.exp(0.2*(beta1 - 0.75)/0.157))
alpha11 = 0.1 - 0.157**2/0.2*(1 - np.exp(0.2*((2*beta1) - 0.75)/0.157))
```

```python
tmax2 = 2000
ts2 = np.linspace(0,200,tmax2)
```

```python
x0 = [0.8,0.5,0.05]
xs = odeint(resident,x0,ts2)
Ss = xs[:,0]
I1s = xs[:,1]
D11s = xs[:,2]

S=Ss[1999]
I1=I1s[1999]
D11=D11s[1999]
```

```python
# Time Course
plt.plot(ts2,Ss)
plt.plot(ts2,I1s)
plt.plot(ts2,D11s)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Numbers', fontsize=14)
plt.xlim((0,200))
plt.ylim((0))
plt.legend(['S', r'$I_1$', r'$D_{11}$'], loc='upper right')
plt.show()
```
