```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%matplotlib inline
```

```python
def resident(x, t):
    lamda = (beta1*(x[1]+x[2]))
    sdot = rho - mu*x[0] - sigmaS*lamda*x[0] + phi1*x[1] + phi11*x[2]
    i1dot = sigmaS*lamda*x[0] - (mu+alpha1+phi1+(sigmaI*lamda))*x[1]
    d11dot = sigmaI*lamda*x[1] - (mu+alpha1+phi11)*x[2]
    return sdot,i1dot,d11dot
```

```python
mu = 0.02
rho = 0.02
sigmaS = 1
sigmaI = 5
phi1 = 0.4
phi11 = 0.2

beta1=0.75
alpha1 = 0.1 - 0.157**2/0.2*(1 - np.exp(0.2*(beta1 - 0.75)/0.157))

phim = 0.45
phi1m = 0.25
```

```python
tmax2 = 5000
ts2 = np.linspace(0,500,tmax2)
```

# $\mu = 0.01$

```python
mu = 0.01
```

```python
x0 = [0.8,0.5,0.05]
xs = odeint(resident,x0,ts2)
Ss = xs[:,0]
I1s = xs[:,1]
D11s = xs[:,2]

S=Ss[4999]
I1=I1s[4999]
D11=D11s[4999]
```

```python
print("S =",S, "I1 =",I1, "D11 =", D11)
```

```python
# Time Course
plt.plot(ts2,Ss)
plt.plot(ts2,I1s)
plt.plot(ts2,D11s)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Numbers', fontsize=14)
plt.xlim((0,500))
plt.ylim((0))
plt.legend(['S', r'$I_1$', r'$D_{11}$'], loc='upper right')
plt.title(r'$\mu = 0.01$')
plt.show()

### S, I & D coexist for these parameter values
```
